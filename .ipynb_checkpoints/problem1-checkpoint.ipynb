```python
# Problem 1
%matplotlib inline
import csv
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

# Counting state visits
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA",
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD",
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ",
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC",
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]

visits = []

with open("visits.csv", "rt") as f:
    reader = csv.reader(f)
    for row in reader:
        visits.append(row[1])

num_visits = {}
for state in states:
    num_visits[state] = visits.count(state)

print(num_visits)

state_repub = {}
# Party affiliaition in 2014 per state
with open("partyaffiliation.csv", "rt") as f:
    reader = csv.reader(f)
    for row in reader:
        state_repub[row[0]] = row[1]

# Proportions for each states
results = []
with open("presidential_results.csv", "rt") as f:
    reader = csv.reader(f)
    currState = "AK"
    currVotesTrump = 0.0
    currVotesTotal = 0.0

    for row in reader:
        #print(currState + " " + row[8])
        #print(currState == row[8])
        if (currState != row[8]):
            print(currState)
            print(currVotesTrump/currVotesTotal)
            results.append([currState, currVotesTrump/currVotesTotal])

            currState = row[8]
            currVotesTrump = 0
            currVotesTotal = 0

        currVotesTrump = currVotesTrump + float(row[2])
        currVotesTotal = currVotesTotal + float(row[3])
```

```python
# Creating logistic model
X = []
Y = []
for states in results: # states is an array [CA, 0.5]
    X.append(num_visits[states[0]])
    Y.append(states[1])

    print(states[0] + " " + str(num_visits[states[0]]) + " " + str(states[1]))

slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)
plt.plot(X, Y, 'o', label="original data")
plt.plot(X, intercept + slope*np.array(X), 'r', label="fitted line")
plt.legend()
plt.show()


```
